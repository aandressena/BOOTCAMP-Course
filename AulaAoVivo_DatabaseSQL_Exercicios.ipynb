```python
consulta_db("SELECT title, rating FROM movies ORDER BY rating LIMIT 10")
```

## Liste todos os filmes de drama. ✓

```python
consulta_db("SELECT m.title, g.name, m.genre_id, g.id FROM movies AS m INNER JOIN genres AS g ON m.genre_id = g.id WHERE g.name = 'Drama'")
```

## Existem quantos filmes de suspense? ✓

```python
consulta_db("SELECT m.title, g.name FROM movies AS m INNER JOIN genres AS g ON m.genre_id = g.id WHERE g.name = 'Thriller'")
```
